The 'players' dataset has:
- 196 observations
- 7 variables:
    - experience, a word representing the player's prior experience with Minecraft, ranging from 'Beginner' to 'Pro'.
    - subscribe, a boolean denoting whether or not the player has subscribed to the game-related newsletter.
    - hashedEmail, a scrambled version of the player's email.
    - played_hours, the amount of time the player has spent on the Minecraft server in hours, rounded to one decimal place. Mean: 5.85
    - name, the player's name.
    - gender, the player's gender.
    - Age, the player's age in years. Mean: 21.14

```r
players = read_csv("data/players.csv")
hours = summarize(players, played_hours = mean(played_hours, na.rm = TRUE))
hours
age = summarize(players, Age = mean(Age, na.rm = TRUE))
age
```
